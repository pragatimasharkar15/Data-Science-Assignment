# Q2 →→→ANS→→→DBSCAN

```python
air = pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
air
```

```python
air.drop(columns=['ID#'] , inplace= True)
air
```

```python
air.describe()
```

```python
air.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(air)
x
```

```python
from sklearn.preprocessing import normalize
```

```python
air = pd.DataFrame(normalize(air) , columns=air.columns)
air
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5  , min_samples=3)

y = db.fit_predict(df)
y
```

```python
from sklearn.metrics import silhouette_score
```

```python
x
```

```python
y
```

```python
silhouette_score(df, y)
```

```python
air[air.duplicated()]
```

```python
air.drop(index=[3876] , inplace= True)
air
```

```python
air['cluster'] = y
air
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors= 5).fit(air)
neigh
```

```python
d ,index = neigh.kneighbors(air)
d
```

```python
distance = np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance graph')
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance graph')

plt.axhline(y = 0.05 , linestyle = '--' , color = 'red')
```

```python
db = DBSCAN(eps=0.05 ,min_samples=5)

yp = db.fit_predict(air)
yp
```

```python
silhouette_score(air , yp)
```
