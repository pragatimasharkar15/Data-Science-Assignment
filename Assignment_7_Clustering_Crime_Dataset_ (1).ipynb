```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x = sc.fit_transform(df)
x
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x, method='complete')
dendrogram = hierarchy.dendrogram(lk)
plt.axhline(y = 5, linestyle = '--', color = 'red')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag = AgglomerativeClustering(n_clusters=2)

ypred = ag.fit_predict(x)
ypred
```

```python
df['cluster'] = ypred
df
```

# Q1 →→→ANS→→→DBSCAN


```python
crime = pd.read_csv('crime_data.csv')
crime
```

```python
crime.drop(columns=['Unnamed: 0'] , inplace= True)
crime
```

```python
crime.describe()
```

```python
crime.boxplot()
```

```python
sns.pairplot(crime)
```

```python
sc = StandardScaler()
x=sc.fit_transform(crime)
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=1 , min_samples=3)

y =db.fit_predict(x)
y
```

```python
from sklearn.metrics import silhouette_score
```

```python
x
```

```python
y
```

```python
silhouette_score(x,y)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=3).fit(x)
neigh
```

```python
d,index = neigh.kneighbors(x)
```

```python
d
```

```python
distance = np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance Graph')
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance Graph')

plt.axhline(y=1.25, linestyle = '--', color = 'red')
```

```python
db = DBSCAN(eps=1.25 ,min_samples=3)
yp = db.fit_predict(x)
yp
```

```python
silhouette_score(x ,yp)
```

```python
crime['cluster']= yp
crime
```

```python
plt.scatter(crime['Murder'], crime['cluster'] ,  cmap='viridis')
```
