```python
#Two Sample t test with two sided cr
#TS = Xnbar - Ynbar/UnderRoot*(2*Sigma*Sigma/n)
u1 = data[["Unit A"]]
u2 = data[["Unit B"]]
from scipy import stats 
zcalc, pval = stats.ttest_ind(u1,u2,alternative = "two-sided")
print("Z-Calculated value = ", zcalc.round(3))
print("P-value = ", pval.round(3))
if(pval<0.05):
    print("H1 shoud get accepted and H0 shoud get rejected")
else:
    print("H0 shoud get accepted and H1 shoud get rejected")
#5% Alpha= 1.64, it is two sided so 2.5%Alpha= +-0.82 CR both sides
#Z-calc(0.723) < +0.82 & > -0.82 -----H0 shoud get accepted
#Pvalue(0.472)>0.05------ 47 % chance that H0 shoud get accepted
```
